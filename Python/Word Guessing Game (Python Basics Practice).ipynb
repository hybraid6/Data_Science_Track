## Word Guessing Game

1. **Task Description**:
Create a game that allows a player to guess a word chosen randomly from a predefined list of words.

2. **Instructions**:
     - **Step 1**: Import the `random` module.
     - **Step 2**: Create a list of words called `words_list` containing several words.
     - **Step 3**: Print a welcome message to the player, informing them about the game and displaying the list of words they can guess from.
     - **Step 4**: Use the `random.choice()` function to select a random word from the `words_list` and assign it to the variable `word_to_guess`.
     - **Step 5**: Create a boolean variable called `guessed_correctly` and set it to `False`. This variable will track whether the player has guessed the word correctly.
     - **Step 6**: Start a `while` loop that continues until `guessed_correctly` is `True`.
     - **Step 7**: Inside the loop, prompt the player to enter their guess using the `input()` function. Strip any leading or trailing whitespace from the input and convert it to lowercase.
     - **Step 8**: Check if the guess matches the `word_to_guess`. If it does, print a congratulatory message informing the player that they have guessed the word correctly and set `guessed_correctly` to `True` to end the loop.
     - **Step 9**: If the guess is incorrect, inform the player that their guess is incorrect and prompt them to guess again.
     - **Step 10**: Continue looping until the player guesses the word correctly.

````python
# Import the random module (module)
import random

# List of words to guess from
words_list = ["python", "java", "kotlin", "javascript", "ruby", "swift"]

# Welcome message
print("Welcome to the Word Guessing Game!")
print("I'm thinking of a word from the following list:")
print(words_list)

# Generate a random word from the list
word_to_guess = random.choice(words_list)

# Variable to track if the correct word has been guessed
guessed_correctly = False

# Continue prompting the player for guesses until they guess correctly
while not guessed_correctly:
    # Prompt the player to enter their guess
    guess = input("Enter your guess: ").strip().lower()
    
    # Check if the guess is correct
    if guess == word_to_guess:
        print("Congratulations! You guessed the word correctly!")
        guessed_correctly = True  # End the loop
    else:
        print("Your guess is incorrect. Guess again.")

````

### Explanation

1. **Import the random module (a library)**:
   ```python
   import random
   ```
   This allows us to use functions from the `random` module, such as selecting a random word.

2. **Define a list of words**:
   ```python
   words_list = ["python", "java", "kotlin", "javascript", "ruby", "swift"]
   ```
   This list contains the words that the player will be guessing from.

3. **Print the welcome message**:
   ```python
   print("Welcome to the Word Guessing Game!")
   print("I'm thinking of a word from the following list:")
   print(words_list)
   ```
   These lines introduce the game to the player and show the list of possible words to be selected.

4. **Select a random word from the list**:
   ```python
   word_to_guess = random.choice(words_list)
   ```
   This line selects a random word from the list for the player to guess and assigns the selection to 'word_to_guess'.

5. **Initialize the guessed_correctly variable**:
   ```python
   guessed_correctly = False
   ```
   This variable tracks whether the player has guessed the word correctly.

6. **Use a while loop to keep asking for guesses until the correct word is guessed**:
   ```python
   '''
   This guessed_correctly part may turn your head a little (It turned mine). 
   A while loop needs a condition to be True to continue running; it stops when the condition becomes False
   guessed_correctly is currently False but the 'not' in the following line turns it to True.
   '''
   while not guessed_correctly:
       # Prompt the player to enter their guess
       guess = input("Enter your guess: ").strip().lower() 
       # The 'input' function creates a box which the recipient is expected to type in.
       # 'strip removes extra spaces and 'lower' changes the input to small letters
       # Check if the guess is correct
       if guess == word_to_guess:
           print("Congratulations! You guessed the word correctly!")
           guessed_correctly = True  # End the loop
       else:
           print("Your guess is incorrect. Guess again.")
   ```
   - The `while` loop continues until `guessed_correctly` is set to `True`.
   - The `input` function prompts the player for their guess.
   - The `strip().lower()` methods are used to remove any leading/trailing spaces and convert the guess to lowercase to make the comparison case-insensitive.
   - The `if` statement checks if the guess matches the selected word.
   - If the guess is correct, a congratulatory message is printed and `guessed_correctly` is set to `True`, ending the loop.
   - If the guess is incorrect, the player is prompted to guess again.
